### 1.3 Generate Necessary Variables

```python
# Select the `Outcome` column and use the `value_counts()` method 
spacex['Outcome'].value_counts()
```

Create a landing outcome label from Outcome column: We convert those outcomes into Training Labels with `1` means the booster successfully landed `0` means it was unsuccessful.

```python
dict(spacex["Outcome"].items())
```

```python
# Assign landing_class values as follows:
# landing_class = 0 if landing unsuccessful
# landing_class = 1 if landing successful

# hint: spacex["Outcome"].items() is the dictionary output above.
landing_class = []
for key, value in spacex["Outcome"].items():
    if 'True' in value:
        landing_class.append(1)
    else:
        landing_class.append(0)
```

```python
# append list to the dataframe
spacex['landing_class'] = landing_class
```

```python
# Then check the data frame using head() method
spacex['landing_class'] = landing_class
spacex.head()
```

```python
# You can also sellect first 20 rows of two columns "Outcome" and "landing_class"
# to check whether the variable was created correctly.
spacex[['Outcome', 'landing_class']].head(20)
```

## 2. Compare Success Rate by Different Groups

```python
# Use value_counts() method to determine the number and 
# occurrence of each orbit in the  column 'Orbit'
spacex[['Outcome', 'landing_class']].head(20)
```

```python
# We see the success rate by orbit using groupby() method

# hint: success rate is mean value of landing_class
spacex.groupby('Orbit')['landing_class'].mean()
```

```python
# Let's see the number of launches for each site.
# Apply value_counts() on column LaunchSite
spacex['LaunchSite'].value_counts()
```

```python
# Let's see and compare the success rate by
# 'LaunchSite' using gourpby()

# hint: success rate is mean value of landing_class
spacex.groupby('LaunchSite')['landing_class'].mean()
```

```python
# Let's see success rate (mean value of landing_class) by Orbit by Reused
spacex.groupby(['Orbit', 'Reused'])['landing_class'].mean()
```

```python
# You can apply groupby() with agg() method to show as a dataframe
spacex.groupby(['Orbit', 'Reused']).agg({'landing_class': 'mean'})
```
